```python
# This file is used to visualize attentions
import torch
```

```python
def vis(attentions):
  import matplotlib.pyplot as plt
  fig, axes = plt.subplots(5, 4, figsize=(12, 12)) # 5 blocks, each 4 heads
  for i, mhblock in enumerate(attentions[0]):
    for j, h in enumerate(mhblock):
      axes[i, j].imshow(h)
      axes[i, j].set_title(f"{j+1} Head")
      axes[i, j].set_xticks([])
      axes[i, j].set_yticks([])

  fig.suptitle('Attention Weights for Each Transformer Block and Head', fontsize=12)
  plt.xlabel('Head')
  plt.ylabel('Transformer Block')

  plt.tight_layout()
  plt.show()
```

```python
attentions = torch.load('attentions.pt')
attentions.shape
vis(attentions)
```
